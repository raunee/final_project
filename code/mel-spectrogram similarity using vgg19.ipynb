```python
feature_vec = []

def img_to_vec(png_name):
  image = cv2.imread(png_path + '/' + png_name)
  image = cv2.resize(image, dsize=(224,224))
  image = img_to_array(image)
  image = image.reshape((1, image.shape[0],image.shape[1],image.shape[2]))
  image = preprocess_input(image)
  vec = model.predict(image)
  feature_vec.append(vec[0])
```

```python
png_list = os.listdir(png_path)
```

```python
from tqdm import tqdm
```

```python
for png_name in tqdm(png_list):
  img_to_vec(png_name)
```

```python
len(feature_vec)
```

```python
labels = png_list
```

```python
cols = [f'vec{i}' for i in range(len(feature_vec[0]))]
```

```python
sim_df = pd.DataFrame(similarity, index=labels, columns=labels)
```

```python
sim_df
```

```python
sim_df.to_csv(base_path+'/music_similarity.csv', sep=',', index = False)
```

```python
def find_similar_songs(name, n=5):
  series = sim_df[name].sort_values(ascending=False)
  music = []
  for a in png_list:
    if a[:-6] == name[:-6]:
      music.append(a)

  series = series.drop(music)

  return series.head(n).to_frame()
```
