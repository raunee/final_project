## trim audio(1차)
- 음원 길이: 30초
- 음원 시작 지점: 20초 간격
- 30초 이하 음원만 남는 경우 마지막 부분을 끝지점으로 하고 그 앞 30초 자름
- 원본 파일이 30초 이하인 경우 모자란 부분에 공백 추가

```python
base_path = '/content/gdrive/My Drive/project'

audio_path = base_path + '/tmp'
save_path = base_path + '/trim'
```

```python
y_blank, sr = librosa.load(base_path + '/blank.wav', sr=44100)
```

```python
def trim_audio_data(audio_file, save_file):
  sr = 44100
  sec = 30

  y, sr = librosa.load(audio_file, sr=sr)

  i = sr*sec
  m = 0
  n = 0
  while True:
    if i + m <= len(y):
      ny = y[m : i + m]
      sf.write(save_file + '_' + str(n) + '.wav', ny, sr)
    else:
      if n == 0:
        ny = np.concatenate([y, y_blank[:i+m-len(y)]])
        sf.write(save_file + '_' + str(n) + '.wav', ny, sr)
      else:
        ny = y[len(y) - i : len(y)]
        sf.write(save_file + '_' + str(n) + '.wav', ny, sr)
      break
    m += sr*20
    n += 1
```

```python
label = []
orig = []
num = []
n = 0
i = 0
save_list = os.listdir(save_path)
for save_file in save_list:
  if save_file[-5] == str(n):
    orig.append(save_file[:-6])
    label.append('music' + str(i+1))
    num.append(n+1)
    n += 1
  else:
    i += 1
    n = 0
    orig.append(save_file[:-6])
    label.append('music' + str(i+1))
    num.append(n+1)
    n += 1
```

```python
import pandas as pd
```

```python
df = pd.DataFrame((zip(save_list, orig, label, num)), columns = ['file name', 'original file', 'label', 'order'])
```
